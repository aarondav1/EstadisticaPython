```
import pandas as pd

df = pd.read_excel("Dataset Proyecto Final Estadistica.xlsx")
df[' H o r a '] = pd.to_datetime(df[' H o r a ']).dt.time
df['F e c h a '] = pd.to_datetime(df['F e c h a '])
df['F e c h a '] = df.apply(lambda r : pd.datetime.combine(r['F e c h a '],r[' H o r a ']),1)
dfVar1 = df[['F e c h a ',' V r m s   p h - n   A N   M e d ']]
dfVar1.set_index('F e c h a ',inplace=True)
dfVar1.plot(grid=True, figsize=(20,10))
```
