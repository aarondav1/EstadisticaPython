```
import pandas as pd

df = pd.read_excel("Dataset Proyecto Final Estadistica.xlsx")
df[' H o r a '] = pd.to_datetime(df[' H o r a ']).dt.time
df['F e c h a '] = pd.to_datetime(df['F e c h a '])
df['F e c h a '] = df.apply(lambda r : pd.datetime.combine(r['F e c h a '],r[' H o r a ']),1)

dfVar1 = df[['F e c h a ',' P o t e n c i a   A c t i v a   A N   M e d ']]
dfVar2 = df[['F e c h a ',' P o t e n c i a   A c t i v a   B N   M e d ']]
dfVar3 = df[['F e c h a ',' P o t e n c i a   A c t i v a   C N   M e d ']]
dfVar4 = df[['F e c h a ',' P o t e n c i a   A c t i v a   T o t a l   M e d ']]
dfVar5 = df[['F e c h a ',' P o t e n c i a   R e a c t i v a   A N   M e d ']]
dfVar6 = df[['F e c h a ',' P o t e n c i a   R e a c t i v a   B N   M e d ']]
dfVar7 = df[['F e c h a ',' P o t e n c i a   R e a c t i v a   C N   M e d ']]
dfVar8 = df[['F e c h a ',' P o t e n c i a   R e a c t i v a   T o t a l   M e d ']]
dfVar9 = df[['F e c h a ',' F a c t o r   d e   P o t e n c i a   A N   M e d ']]
dfVar10 = df[['F e c h a ',' F a c t o r   d e   P o t e n c i a   B N   M e d ']]
dfVar11 = df[['F e c h a ',' F a c t o r   d e   P o t e n c i a   C N   M e d ']]
dfVar12 = df[['F e c h a ',' F a c t o r   d e   P o t e n c i a   T o t a l   M e d ']]

dfVar1.set_index('F e c h a ',inplace=True)
dfVar2.set_index('F e c h a ',inplace=True)
dfVar3.set_index('F e c h a ',inplace=True)
dfVar4.set_index('F e c h a ',inplace=True)
dfVar5.set_index('F e c h a ',inplace=True)
dfVar6.set_index('F e c h a ',inplace=True)
dfVar7.set_index('F e c h a ',inplace=True)
dfVar8.set_index('F e c h a ',inplace=True)
dfVar9.set_index('F e c h a ',inplace=True)
dfVar10.set_index('F e c h a ',inplace=True)
dfVar11.set_index('F e c h a ',inplace=True)
dfVar12.set_index('F e c h a ',inplace=True)

dfVar1.plot(grid=True, figsize=(20,10))
dfVar2.plot(grid=True, figsize=(20,10))
dfVar3.plot(grid=True, figsize=(20,10))
dfVar4.plot(grid=True, figsize=(20,10))
dfVar5.plot(grid=True, figsize=(20,10))
dfVar6.plot(grid=True, figsize=(20,10))
dfVar7.plot(grid=True, figsize=(20,10))
dfVar8.plot(grid=True, figsize=(20,10))
dfVar9.plot(grid=True, figsize=(20,10))
dfVar10.plot(grid=True, figsize=(20,10))
dfVar11.plot(grid=True, figsize=(20,10))
dfVar12.plot(grid=True, figsize=(20,10))
```
