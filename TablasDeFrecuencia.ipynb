```
import pandas as pd

df = pd.read_excel("Dataset Proyecto Final Estadistica.xlsx")

def tablaFrecuencias(df, variable):
  lis = df.unique()
  dat = pd.DataFrame(lis, columns=[variable])
  #Tabla de frecuencias absolutas
  datafi = pd.crosstab(index=df, columns = "fi")
  li = datafi.values
  dat["fi"] = li
  #Tabbla de frecuencia relativa
  datahi = 100 * datafi["fi"] / 5760
  datahi = datahi.values
  dat["hi"] = datahi
  #Obtenemos las frecuencias absolutas acumuladas
  Fi = dat["fi"].values
  a = []
  b = 0
  for c in Fi:
    b = c + b
    a.append(b)
  dat["Fi"] = a
  #Recorremos lista para obtener la frecuencia relativa acumulada
  Hi = dat["hi"].values
  a = []
  b = 0
  for c in Hi:
    b = c + b
    a.append(b)
  dat["Hi"] = a
  print(dat)

tablaFrecuencias(df[' P o t e n c i a   A c t i v a   A N   M e d '], "Potencia Activa AN Med")
tablaFrecuencias(df[' P o t e n c i a   A c t i v a   B N   M e d '], "Potencia Activa BN Med")
tablaFrecuencias(df[' P o t e n c i a   A c t i v a   C N   M e d '], "Potencia Activa CN Med")
tablaFrecuencias(df[' P o t e n c i a   A c t i v a   T o t a l   M e d '], "Potencia Activa Total Med")
tablaFrecuencias(df[' P o t e n c i a   R e a c t i v a   A N   M e d '], "Potencia Reactiva AN Med")
tablaFrecuencias(df[' P o t e n c i a   R e a c t i v a   B N   M e d '], "Potencia Reactiva BN Med")
tablaFrecuencias(df[' P o t e n c i a   R e a c t i v a   C N   M e d '], "Potencia Reactiva CN Med")
tablaFrecuencias(df[' P o t e n c i a   R e a c t i v a   T o t a l   M e d '], "Potencia Reactiva Total Med")
tablaFrecuencias(df[' F a c t o r   d e   P o t e n c i a   A N   M e d '], "Factor De Potencia AN Med")
tablaFrecuencias(df[' F a c t o r   d e   P o t e n c i a   B N   M e d '], "Factor De Potencia BN Med")
tablaFrecuencias(df[' F a c t o r   d e   P o t e n c i a   C N   M e d '], "Factor De Potencia CN Med")
tablaFrecuencias(df[' F a c t o r   d e   P o t e n c i a   T o t a l   M e d '], "Factor De Potencia Total Med")

```
