```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("Dataset Proyecto Final Estadistica.xlsx")

def historigrama(df, variable):
  plt.figure(figsize=(18,8))
  plt.hist(df)
  plt.xlabel(variable)
  plt.ylabel("Frecuencia absoluta")
  plt.show()

historigrama(df[' P o t e n c i a   A c t i v a   A N   M e d '], "Potencia Activa AN Med")
historigrama(df[' P o t e n c i a   A c t i v a   B N   M e d '], "Potencia Activa BN Med")
historigrama(df[' P o t e n c i a   A c t i v a   C N   M e d '], "Potencia Activa CN Med")
historigrama(df[' P o t e n c i a   A c t i v a   T o t a l   M e d '], "Potencia Activa Total Med")
historigrama(df[' P o t e n c i a   R e a c t i v a   A N   M e d '], "Potencia Reactiva AN Med")
historigrama(df[' P o t e n c i a   R e a c t i v a   B N   M e d '], "Potencia Reactiva BN Med")
historigrama(df[' P o t e n c i a   R e a c t i v a   C N   M e d '], "Potencia Reactiva CN Med")
historigrama(df[' P o t e n c i a   R e a c t i v a   T o t a l   M e d '], "Potencia Reactiva Total Med")
historigrama(df[' F a c t o r   d e   P o t e n c i a   A N   M e d '], "Factor De Potencia AN Med")
historigrama(df[' F a c t o r   d e   P o t e n c i a   B N   M e d '], "Factor De Potencia BN Med")
historigrama(df[' F a c t o r   d e   P o t e n c i a   C N   M e d '], "Factor De Potencia CN Med")
historigrama(df[' F a c t o r   d e   P o t e n c i a   T o t a l   M e d '], "Factor De Potencia Total Med")

```
