```
import pandas as pd
import numpy as np

df = pd.read_excel("Dataset Proyecto Final Estadistica.xlsx")

def Medidas(df, variable):
  media = df.mean()
  mediana = df.median()
  moda = df.mode()
  rango = df.max() - df.min()
  iqr = df.quantile(0.75) - df.quantile(0.25)
  var = df.var(ddof=0)
  std = np.sqrt(var)
  cdv = std / df.mean()
  q1 = df.quantile(0.25)
  q2 = df.quantile(0.5)
  q3 = df.quantile(0.75)
  max = df.max()
  min = df.min()
  print("""
      *** Variable: %s ***
      MEDIDAS DE TENDENCIA CENTRAL
      Media: %f
      Mediana: %f
      Moda: %f

      MEDIDAS DE VARIACIÓN
      Rango: %f
      Rango Intercuartil: %f
      Desviacion Estandar: %f
      Varianza: %f
      Coeficiente de variacion: %f

      MEDIDAS DE POSICIÓN RELATIVA
      Cuartil Q1: %f
      Cuartil Q2: %f
      Cuartil Q3: %f
      Maximo: %f
      Minimo: %f
      """ % (variable,media, mediana, moda, rango, iqr, std, var, cdv, q1, q2, q3, max, min))
Medidas(df[' P o t e n c i a   A c t i v a   A N   M e d '], "Potencia Activa AN Med")
Medidas(df[' P o t e n c i a   A c t i v a   B N   M e d '], "Potencia Activa BN Med")
Medidas(df[' P o t e n c i a   A c t i v a   C N   M e d '], "Potencia Activa CN Med")
Medidas(df[' P o t e n c i a   A c t i v a   T o t a l   M e d '], "Potencia Activa Total Med")
Medidas(df[' P o t e n c i a   R e a c t i v a   A N   M e d '], "Potencia Reactiva AN Med")
Medidas(df[' P o t e n c i a   R e a c t i v a   B N   M e d '], "Potencia Reactiva BN Med")
Medidas(df[' P o t e n c i a   R e a c t i v a   C N   M e d '], "Potencia Reactiva CN Med")
Medidas(df[' P o t e n c i a   R e a c t i v a   T o t a l   M e d '], "Potencia Reactiva Total Med")
Medidas(df[' F a c t o r   d e   P o t e n c i a   A N   M e d '], "Factor De Potencia AN Med")
Medidas(df[' F a c t o r   d e   P o t e n c i a   B N   M e d '], "Factor De Potencia BN Med")
Medidas(df[' F a c t o r   d e   P o t e n c i a   C N   M e d '], "Factor De Potencia CN Med")
Medidas(df[' F a c t o r   d e   P o t e n c i a   T o t a l   M e d '], "Factor De Potencia Total Med")
```
