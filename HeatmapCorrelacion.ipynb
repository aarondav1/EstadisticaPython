```
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_excel("Dataset Proyecto Final Estadistica.xlsx")

dfa = df[[' P o t e n c i a   A c t i v a   A N   M e d ',' P o t e n c i a   A c t i v a   B N   M e d ',' P o t e n c i a   A c t i v a   C N   M e d ',
          ' P o t e n c i a   A c t i v a   T o t a l   M e d ',' P o t e n c i a   R e a c t i v a   A N   M e d ',' P o t e n c i a   R e a c t i v a   B N   M e d ',
          ' P o t e n c i a   R e a c t i v a   C N   M e d ',' P o t e n c i a   R e a c t i v a   T o t a l   M e d ',
          ' F a c t o r   d e   P o t e n c i a   A N   M e d ',' F a c t o r   d e   P o t e n c i a   B N   M e d ',
          ' F a c t o r   d e   P o t e n c i a   C N   M e d ',' F a c t o r   d e   P o t e n c i a   T o t a l   M e d ']]

def show_heatmap(data):
    plt.matshow(data.corr())
    plt.xticks(range(data.shape[1]), data.columns, fontsize=14, rotation=90)
    plt.gca().xaxis.tick_bottom()
    plt.yticks(range(data.shape[1]), data.columns, fontsize=14)

    cb = plt.colorbar()
    cb.ax.tick_params(labelsize=14)
    plt.title("Feature Correlation Heatmap", fontsize=14)
    plt.show()

show_heatmap(dfa)
```
